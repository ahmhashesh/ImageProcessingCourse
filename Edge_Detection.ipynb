```python
import numpy as np
import cv2
%matplotlib inline
from matplotlib import pyplot as plt
img = cv2.imread('files/fig/Lenna.png',0)
edges = cv2.Canny(img,100,200)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

# Sobel:

reference code https://gist.github.com/rahit/c078cabc0a48f2570028bff397a9e154

```python
img_gaussian = cv2.GaussianBlur(img,(3,3),0)
img_sobelx = cv2.Sobel(img_gaussian,cv2.CV_8U,1,0,ksize=5)
img_sobely = cv2.Sobel(img_gaussian,cv2.CV_8U,0,1,ksize=5)
img_sobel = img_sobelx + img_sobely
plt.subplot(131)
plt.imshow(img_sobelx,cmap = 'gray'),plt.title("Sobel X")
plt.subplot(132)
plt.imshow(img_sobely,cmap = 'gray'),plt.title("Sobel Y")
plt.subplot(133)
plt.imshow(img_sobel,cmap = 'gray'),plt.title("Sobel")
```
